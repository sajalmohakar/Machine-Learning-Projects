```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv("/content/breastdata.csv")
```

```python
data.head()
```

```python
data.replace({"diagnosis":{'M':0,'B':1}},inplace=True)
```

```python
data.head()
```

```python
data.describe()
```

```python
data.drop(['id'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data = data.drop(['Unnamed: 32'],axis=1,inplace=True)
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
model = LogisticRegression()
```

```python
x= data.drop(['diagnosis'],axis=1)
y= data['diagnosis']
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x.head()
```

```python
X = scaler.transform(x)
```

```python
print(X)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred = model.predict(x_train)
```

```python
accuracy = accuracy_score(x_train_pred,y_train)
print(accuracy)
```

```python
x_test_pred = model.predict(x_test)
```

```python
accuracy1 = accuracy_score(x_test_pred,y_test)
print(accuracy1)
```

```python
from sklearn import svm
```

```python
svm_model = svm.SVC(kernel='linear')
```

```python
svm_model.fit(x_train,y_train)
```

```python
x_train_pred = svm_model.predict(x_train)
```

```python
accuracy = accuracy_score(y_train,x_train_pred)
print(accuracy)
```

```python
x_test_pred = svm_model.predict(x_test)
```

```python
accuracy= accuracy_score(y_test,x_test_pred)
print(accuracy)
```
