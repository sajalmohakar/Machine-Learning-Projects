```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("/content/asthma_disease_data.csv")
```

```python
print(data)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data = data.drop(['PatientID','Age'],axis=1)
```

```python
data.isnull().sum()
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
data.head()
```

```python
data = data.drop(['DoctorInCharge'],axis=1)
```

```python
data.head()
```

```python
x = data.drop(['Diagnosis'],axis=1)
y = data['Diagnosis']
```

```python
y.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
model = KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
x_train_pred = model.predict(x_train)
```

```python
print(x_train_pred)
```

```python
accuracy=accuracy_score(y_train,x_train_pred)
print(accuracy)
```

```python
x_test_pred = model.predict(x_test)
accuracy= accuracy_score(y_test,x_test_pred)
print(accuracy)
```
